```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
data = pd.read_csv("healthcare_dataset_cleaned_labelled.csv")

print("Dataset Shape:", data.shape)

data.head()
```

```python
print("Dataset Information")

data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(data["Age"], bins=20, kde=True)
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Number of Patients")

plt.show()
```

```python
plt.figure(figsize=(7,5))

order = data["Gender"].value_counts().index

ax = sns.countplot(
    x="Gender",
    data=data,
    order=order,
    hue="Gender",
    palette="Set2",
    legend=False
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Distribution of Patients by Gender", fontsize=14)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Number of Patients", fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(10,5))

order = data["Medical Condition"].value_counts().index

ax = sns.countplot(
    y="Medical Condition",
    data=data,
    order=order,
    hue="Medical Condition",
    palette="viridis",
    legend=False
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Distribution of Patients by Medical Condition")
plt.xlabel("Number of Patients")
plt.ylabel("Medical Condition")

plt.show()
```

```python
plt.figure(figsize=(9,5))

order = data["Blood Type"].value_counts().index

ax = sns.countplot(
    x="Blood Type",
    data=data,
    order=order,
    hue="Blood Type",
    palette="Set2",
    legend=False
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Distribution of Patients by Blood Type", fontsize=14)
plt.xlabel("Blood Type")
plt.ylabel("Number of Patients")

plt.show()
```

```python
plt.figure(figsize=(8,5))

order = data["Admission Type"].value_counts().index

ax = sns.countplot(
    x="Admission Type",
    data=data,
    order=order,
    hue="Admission Type",
    palette="Set2",
    legend=False
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Distribution of Patients by Admission Type", fontsize=14)
plt.xlabel("Admission Type")
plt.ylabel("Number of Patients")

plt.show()
```

```python
plt.figure(figsize=(8,5))

order = data["Test Results"].value_counts().index

ax = sns.countplot(
    x="Test Results",
    data=data,
    order=order,
    hue="Test Results",
    palette="Set2",
    legend=False
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Distribution of Test Results", fontsize=14)
plt.xlabel("Test Result")
plt.ylabel("Number of Patients")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(data["Billing Amount"], bins=30, kde=True)

plt.title("Billing Amount Distribution")

plt.xlabel("Billing Amount")

plt.show()
```

```python
plt.figure(figsize=(10,6))

ax = sns.boxplot(
    x="Medical Condition",
    y="Hospital Stay",
    data=data,
)

plt.title("Hospital Stay by Medical Condition", fontsize=14)
plt.xlabel("Medical Condition", fontsize=12)
plt.ylabel("Hospital Stay (Days)", fontsize=12)

plt.xticks(rotation=15)

plt.show()
```

```python
plt.figure(figsize=(9,5))

ax = sns.boxplot(
    x="Admission Type",
    y="Billing Amount",
    data=data,
    hue="Admission Type",
    palette="Set2",
    legend=False
)

plt.title("Billing Amount by Admission Type", fontsize=14)
plt.xlabel("Admission Type", fontsize=12)
plt.ylabel("Billing Amount", fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(12,8))

columns = [
    "Age",
    "Billing Amount",
    "Hospital Stay",
    "Gender Label",
    "Blood Type Label",
    "Admission Type Label",
    "Medical Condition Label",
    "Test Results Label"
]

correlation = data[columns].corr()

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5,
    square=True
)

plt.title("Correlation Between Numerical Features", fontsize=15)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.regplot(
    x="Age",
    y="Billing Amount",
    data=data,
    scatter_kws={"alpha":0.08, "s":8},
    line_kws={"color":"red"}
)

plt.title("Age vs Billing Amount")
plt.xlabel("Age (Years)")
plt.ylabel("Billing Amount")

plt.show()
```

```python
plt.figure(figsize=(10,6))

avg_bill = data.groupby("Medical Condition")["Billing Amount"].mean().sort_values(ascending=False)

ax = sns.barplot(
    x=avg_bill.index,
    y=avg_bill.values,
    hue=avg_bill.index,
    palette="viridis",
    legend=False
)

for container in ax.containers:
    ax.bar_label(container, fmt="%.0f")

plt.title("Average Billing Amount by Medical Condition", fontsize=14)
plt.xlabel("Medical Condition")
plt.ylabel("Average Billing Amount")

plt.xticks(rotation=15)

plt.show()
```
